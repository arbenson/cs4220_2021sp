```julia
using Plots
using Roots
```

```julia
g1(x) = (x + 1) * (x - 1)^8 - 1e-8
g2(x) = (x + 1)^(1/8) * (x - 1) - 1e-1
```

```julia
xs = range(0.75, step=0.001, stop=1.25)
v1 = [g1(x) for x in xs]
v2 = [g2(x) for x in xs]
;
```

```julia
plot(xs,  v1, label="g1")
plot!(xs, v2, label="g2")
```

```julia
find_zero(g1, (0.75, 1.25), Bisection())
```

```julia
v = find_zero(g2, (0.75, 1.25), Bisection())
```

```julia
g1(v), g2(v)
```
